```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from fastbook import *
```

```python
def loss(w):
      return w**2
```

```python
plot_function(loss, 'w', 'loss', min = -5, max = 5)
```

```python
plot_function(loss, 'w', 'loss', min=-5, max=5)
plt.scatter(-3, loss(-3), color = 'red')
```

```python
# Parabola function y = x^2
def f(x):
    return x**2

# Analitic hosila: f'(x) = 2x
def slope(x):
    return 2 * x

# Tangent line: y = m*(x - x1) + y1
def tangent_line(x, x1, y1):
    m = slope(x1)
    return m * (x - x1) + y1

# Grafic limits
x = np.linspace(-5, 5, 400)

# Tangent pointing 
x1 = 1
y1 = f(x1)
x_tan = np.linspace(x1 - 1, x1 + 1, 100)
y_tan = tangent_line(x_tan, x1, y1)

# Grafic plot
plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label='Parabola: y = x²')
plt.scatter([x1], [y1], color='red', label=f'Tangent point (x₁={x1})')
plt.plot(x_tan, y_tan, '--', color='green', linewidth=2, label='Tangent line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola and its Tangent Line')
plt.legend()
plt.grid(True)
plt.show()
```
