```python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
%matplotlib inline
```

=> Artifitial data

```python
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]



def classic_program(x): return x*70
       
y_pred = [classic_program(x) for x in x_data]

# plotting

plt.figure(figsize=(6, 5))
plt.scatter(x_data, y_data, label = 'Actual data set')
plt.plot(x_data, y_pred, 'r', label = 'Program')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('x*70')
plt.show()
```

=> to add Wight assignment 1 and 2 method

```python
data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]



def weight_assignment_program(x, w): return w*x*70

w = 1.4

y_pred = [weight_assignment_program(x, w) for x in x_data]

# plotting

plt.figure(figsize=(6, 5))
plt.scatter(x_data, y_data, label = 'Actual data set')
plt.plot(x_data, y_pred, 'r', label = 'weighted Program')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title(f'{w}*x*70')
plt.show()
```

=> 0 train Machine Learning 3-method

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as pyplot
%matplotlib inline
```

```python
# data

data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# regression

regression = LinearRegression()
regression.fit(x_data, y_data)

y_pred = regression.predict(x_data)



# plotting

plt.figure(figsize=(6, 5))
plt.scatter(x_data, y_data, label = 'Actual data set')
plt.plot(x_data, y_pred, 'r', label = 'Machine Learning Program')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
equation = f'{regression.coef_[0]:.3f}x+{regression.intercept_:.3f}'
plt.title(equation)
plt.show()
```

=> Deep Learing model for this model

```python
from sklearn.datasets import make_regression
from sklearn.neural_network import MLPRegressor
from matplotlib import pyplot as pyplot
%matplotlib inline
```

```python
# data

data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)
x_data, y_data = data[0], data[1]

# regression

regression = MLPRegressor(max_iter=2000)
regression.fit(x_data, y_data)

y_pred = regression.predict(x_data)



# plotting

plt.figure(figsize=(6, 5))
plt.scatter(x_data, y_data, label = 'Actual data set')
plt.plot(x_data, y_pred, 'r', label = 'Deep Learning Program')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
